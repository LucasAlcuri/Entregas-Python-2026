```python
def spoilers(review:str):

    review_split = review.split()
    

    lista_spoilers = input("Ingrese las palabras que desea censurar (separado por comas)")
    lista_spoilers = lista_spoilers.split(", ")
    
    for i in range(len(lista_spoilers)):
        lista_spoilers[i] = lista_spoilers[i].lower()
   
    review_censurado = ""

    for i in range(len(review_split)):
        word = review_split[i]
        word = word.lower()
        if word in lista_spoilers:
            review_split[i] = "*" * len(word)
       
        if i % 12 == 0:
            review_censurado += "\n"
        review_censurado += " " + review_split[i]

    print(review_censurado)   
    return(review_censurado)      
    
```

```python
review = """La película sigue a un grupo de astronautas que
viajan a Marte
en una misión de rescate. El capitán Torres lidera al equipo
a través
de tormentas solares y fallos en el sistema de navegación. Al
llegar
a Marte descubren que la base está abandonada y los
suministros
destruidos. Torres decide sacrificar la nave nodriza para
salvar
al equipo y logran volver a la Tierra en una cápsula de
emergencia.
El final revela que Torres sobrevivio gracias a un pasaje
secreto."""
review_censurado = spoilers(review)
```
