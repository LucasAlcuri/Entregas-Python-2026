```python
def tabla_competidores(rounds:list[dict]):
    
    tabla_cocineros = []
    
    for cocinero in rounds[0]["scores"]:
        diccionario = {}
        diccionario["Nombre"] = cocinero
        diccionario["Puntaje_total"] = 0
        diccionario["Rondas_Ganadas"] = 0
        diccionario["Mejor_ronda"] = 0
        diccionario["Promedio"] = 0
        tabla_cocineros.append(diccionario)
        
        
    return tabla_cocineros
```

```python
def encontrar_pos(tabla_cocineros:list[dict],cocinero:str):
    pos = 0
    for diccionario in  tabla_cocineros:
        
        if diccionario["Nombre"] == cocinero : return pos

        else: pos += 1 
```

```python
def ordenar_tabla(tabla_cocineros:list[dict]):
    
    for i in range(len(tabla_cocineros)):
        
        for j in range(len(tabla_cocineros) - 1 ):
            
            if tabla_cocineros[j]["Puntaje_total"] < tabla_cocineros[j+1]["Puntaje_total"]:
                aux = tabla_cocineros[j]
                tabla_cocineros[j] = tabla_cocineros[j+1]
                tabla_cocineros[j+1] = aux
```

```python
def calcular_promedios(tabla_cocineros:list[dict],rondas:int):
    for diccionario in tabla_cocineros:
        diccionario["Promedio"] = diccionario["Puntaje_total"] / rondas
```

```python
def imprimir_tabla(tabla_cocineros:list[dict],ronda:int):

    if ronda == 6:
        print(f"Tabla posiciones Final de la competencia!\n")
        print(f"{'Cocinero':<12} {'Puntaje':<10} {'Rondas Ganadas':<18} {'Mejor Ronda':<15} {'Promedio':<10}")
    else:
        print(f"Tabla posiciones despues de la ronda {ronda}:\n")
        print(f"{'Cocinero':<12} {'Puntaje':<10} {'Rondas Ganadas':<18}")
    
    print("-" * 70)
    for c in tabla_cocineros:
        if ronda == 6:
            print(f"{c['Nombre']:<12} {c['Puntaje_total']:<10} {c['Rondas_Ganadas']:<18} {c['Mejor_ronda']:<15} {c['Promedio']:.2f}")
        else:
            print(f"{c['Nombre']:<12} {c['Puntaje_total']:<10} {c['Rondas_Ganadas']:<18}") 
```

```python
def competencia(rounds:list[dict],tabla_cocineros:list[dict]):
    
    ronda = 1
   
    for tema in rounds: # tema = {theme,scores}
        puntaje_ganador = -1
        participante_ganador = ""
       
        for participantes in tema["scores"]: # Accedo al diccionario de jueces de
                                             # cada cocinero
            puntaje_total = 0
            participante_actual = participantes

            pos = encontrar_pos(tabla_cocineros,participante_actual)

            for judge in tema["scores"][participante_actual]: # Accedo al puntaje de cada juez y lo sumo
                
                puntaje_total += tema["scores"][participante_actual][judge]
    
            
            if puntaje_total > puntaje_ganador: 

                puntaje_ganador = puntaje_total
                participante_ganador = participante_actual

            if puntaje_total > tabla_cocineros[pos]["Mejor_ronda"]: 
                tabla_cocineros[pos]["Mejor_ronda"] = puntaje_total

            tabla_cocineros[pos]["Puntaje_total"] += puntaje_total    

        
        print(f"""
              Ronda {ronda} - {tema["theme"]:} 
              Ganador: {participante_ganador} ({puntaje_ganador} puntos)
                    """)
        pos = encontrar_pos(tabla_cocineros,participante_ganador)
        tabla_cocineros[pos]["Rondas_Ganadas"] += 1
        
        
        ordenar_tabla(tabla_cocineros)
        imprimir_tabla(tabla_cocineros,ronda)
        ronda += 1

    # Fin competencia
    
    calcular_promedios(tabla_cocineros,ronda-1)
    print("\n" * 2)
    imprimir_tabla(tabla_cocineros,ronda)    
```
