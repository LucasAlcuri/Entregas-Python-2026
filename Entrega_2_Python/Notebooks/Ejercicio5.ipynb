```python
def calcular_precio(zona:tuple,peso:float):
    precio = 0
    if peso <= 1:
        precio = zona[0]
    elif peso  > 1 and peso <= 5:
        precio = zona[1]
    else: precio = zona[2]

    print(f"Costo Envio: {precio}")       
    
```

```python
local = (500, 1000, 2000)
regional = (1000, 2500, 5000)
nacional = (2000, 4500, 8000)

peso = float(input ("Ingrese el peso (kg)  (0 para finalizar)"))


while peso != 0:
    zona = input("Ingrese la zona de destino (local/regional/nacional):")
    zona = zona.lower()
    if zona not in ("local","regional","nacional"):
        print("Zona no válida. Las zonas disponibles son: local, regional, nacional.")
        print()
    else: 
        if zona == "local" : calcular_precio(local,peso)
        
        elif zona == "regional": calcular_precio(regional,peso)
        
        else: calcular_precio(nacional,peso)

    peso = float(input ("Ingrese el peso (kg)"))   
```
