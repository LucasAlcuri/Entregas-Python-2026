```python
def validar_mail(mail:str):
    
    if mail.startswith("@") or mail.endswith("@") or mail.startswith(".") or mail.endswith("."): return False
    
    if mail.count("@") != 1: return False

    mail_split = mail.split("@")

    if mail_split[1].count(".") < 1: return False

    mail_split = mail_split[1].split(".")

    if len(mail_split[len(mail_split)-1]) < 2: return False

    return True
```

```python

#Valide una dirección de email ingresada por el usuario con los siguientes
#criterios:
#Contiene exactamente un @.
#Tiene al menos un carácter antes del @.
#Tiene al menos un punto (.) después del @.
#No empieza ni termina con @ ni con ..
#La parte después del último punto tiene al menos 2 caracteres (el dominio).

mail = input("Ingrese el mail a validar (<FIN> para finalizar)")
while mail != "FIN":
    if validar_mail(mail):
        print("Mail valido")
        mail = input("Ingrese el mail a validar (<FIN> para finalizar)")
    else:
        print("Mail Invalido") 
        mail = input("Ingrese el mail a validar (<FIN> para finalizar)")   
```
