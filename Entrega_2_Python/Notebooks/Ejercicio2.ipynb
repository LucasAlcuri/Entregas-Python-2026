```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
```

```python

max = -1
min = 9999

total_segundos = 0

cancion_mas_larga = {}
cancion_mas_corta = {}

for element in playlist:
    
    num = element["duration"]
    num = num.split(":") 

    min_act = int(num[0])
    segs_act = int(num[1])

    aux = (min_act * 60) + segs_act # Transformo todo a segundos
    total_segundos += aux

    # Maximo
    if aux > max:
        max = aux
        cancion_mas_larga = element
    
    # Minimo
    if aux < min:
        min = aux
        cancion_mas_corta = element

total_min = total_segundos // 60
total_segs = total_segundos % 60
    
print(f"""
      Duracion total: {total_min}M {total_segs}S
      Cancion mas larga: {cancion_mas_larga["title"]} ({cancion_mas_larga["duration"]})       
      Cancion mas corta: {cancion_mas_corta["title"]} ({cancion_mas_corta["duration"]})
    """)  
```
